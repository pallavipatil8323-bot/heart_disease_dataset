```python
# featurtes and taregt separation
#features
x=df.drop(columns='target',axis=1)
#taregt
y=df['target']
```

```python
x.columns
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100,stratify=y)
```

```python
#feature scaling- preprocessing
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
#here we will use logistic regression cuz it is a binary classification problem
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train_scaled,y_train)
```

```python
#predictions
y_pred_train=model.predict(x_train_scaled)
y_pred_test=model.predict(x_test_scaled)
```

```python
#score
model.score(x_train_scaled,y_train)
```

```python
model.score(x_test_scaled,y_test)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred_test)
cm
```

```python
#classification report
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred_test)
print(cr)
```

```python
#decision tree model
#we will check with the decision tree model and compare the result
#This model learns non-linear relationship handles feature interaction better than logistic regression but sometimes tends to overfit

from sklearn.tree import DecisionTreeClassifier
model_dt=DecisionTreeClassifier()
model_dt.fit(x_train,y_train)
```

```python
#predicting
y_pred_train_dt=model_dt.predict(x_train)
y_pred_test_dt=model_dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

#evaluating
acc_train_dt=accuracy_score(y_train,y_pred_train_dt)
acc_test_dt=accuracy_score(y_test,y_pred_test_dt)
print(acc_train_dt)
print(acc_test_dt)
```

```python
#confusion matrix
print(confusion_matrix(y_test,y_pred_test_dt))
#classification report
print(classification_report(y_test,y_pred_test_dt))
```
